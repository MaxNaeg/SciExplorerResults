```python
import os
import sys
import sciexplorer
import json
from quantum_gs_final_runs.physics_run_quantumspins import models
# Set path to examples folder
EXAMPLES_PATH="/Users/mnaegel/Documents/sciexplorer_framework/sciexplorer/examples"
sys.path.append(EXAMPLES_PATH)
from simulators.wave_sim_v2 import predefined_wave_systems
from simulators.wave_sim_v2 import WaveSimulator
import jax.numpy as jnp
import matplotlib.pyplot as plt
import matplotlib
```

```python
# get list of a json files in subdirectorires
exp_files = []
for root, dirs, files in os.walk('./waves_final_runs/'):
    for file in files:
        p = os.path.join(root, file)
        if '/run_' in p:
            exp_files.append(p)
```

```python
time_list = []
gauss_small = 'x_rescaled = x/jnp.max(x)\nphi0=0.1*jnp.exp(-0.5*(x_rescaled**2/0.01))'
gauss_medium = 'x_rescaled = x/jnp.max(x)\nphi0=jnp.exp(-0.5*(x_rescaled**2/0.01))'
two_gauss = 'x_rescaled = x/jnp.max(x)\nphi0=0.2*jnp.exp(-0.5*(x_rescaled**2/0.01)) + 0.7*jnp.exp(-0.5*((x_rescaled-0.3)**2/0.01))'
gauss_mom = 'x_rescaled = x/jnp.max(x)\nphi0=jnp.exp(-0.5*((x_rescaled+0.5)**2/0.01)) * jnp.exp(1j*5*(x_rescaled+0.5))'
plot_evolution = True
init_code  = gauss_medium

font = "Helvetica Neue"
plt.rcParams.update({
    "text.usetex": False,   # Let Matplotlib handle fonts
    "font.family": "sans-serif",
    "font.sans-serif": [font,]
})


for i, f in enumerate(exp_files):
    arguments = None
    with open(f, 'r') as file:
        exp_data = json.load(file)
    for msg in exp_data['messages']:
        tool_calls = msg.get('tool_calls', [])
        for call in tool_calls:
            function_name = call.get('function', {}).get('name', '')
            if function_name == 'save_result_find_eom':
                arguments = json.loads(call.get('function', {}).get('arguments', {}))
    exp_type = f.split('/')[-2]
    #print(f"File {i}/{len(exp_files)}: {f}, Hamiltonian: {arguments.get('Hamiltonian', '')}")
    func_true = predefined_wave_systems[exp_type][0]
    init_true=dict(x_max=5, N_x=100, t_max=20., N_t=200, rhs_choice=exp_type, 
                        rhs_code=None, description_type="FIELD")
    sim_true = WaveSimulator(**init_true)
    func_pred = arguments.get('code', '')
    init_pred = dict(x_max=5, N_x=100, t_max=20., N_t=200,rhs_choice=None, rhs_code=func_pred, description_type="FIELD")
    sim_pred = WaveSimulator(**init_pred)



    print(f"\n{exp_type}------------------------------------------------")
    print(f"R2 {exp_data['result']['R2']}, file: {f}")
    print(f"func_true--------------------")
    print(f"{func_true}")
    print(f"func_pred--------------------")
    print(f"{func_pred}")


    if plot_evolution:
        true_evol_result = sim_true.run_field_evolution_experiment(init_code)
        pred_evol_result = sim_pred.run_field_evolution_experiment(init_code)
        true_evol_result['phis'] /= jnp.max(true_evol_result['phis'])
        pred_evol_result['phis'] /= jnp.max(pred_evol_result['phis'])

        fig, axs = plt.subplots(1, 2, figsize=(8, 3))
        ax = axs[0]
        im1 = ax.imshow(jnp.abs(true_evol_result['phis'])**2, cmap=matplotlib.colormaps['plasma'],extent=[-5, 5, 20, 0])
        ax.set_aspect(.5)
        ax.set_xticks([-5,  0,  5])
        ax.set_yticks([0, 10, 20])
        ax.tick_params(labelleft=False, labelbottom=False, direction='out', bottom=False, top=True)


        ax = axs[1]
        im2 = ax.imshow(jnp.abs(pred_evol_result['phis'])**2, cmap=matplotlib.colormaps['plasma'],extent=[-5, 5, 20, 0])
        ax.set_aspect(.5)
        ax.set_xticks([-5,  0,  5])
        ax.set_yticks([0, 10, 20])
        ax.tick_params(labelleft=False, labelbottom=False, direction='out', bottom=False, top=True)

        cbar_ax = fig.add_axes([0.88, 0.1, 0.04, 0.8])
        fig.colorbar(im2, cax=cbar_ax, ticks=[])  
        fig.tight_layout()
        fig.subplots_adjust(wspace=-0.4, hspace=0.1)
        os.makedirs("./wave_comp_true_exp", exist_ok=True)
        plt.savefig(f"./wave_comp_true_exp/wave_{exp_type}_R2_{exp_data['result']['R2']}.pdf", dpi=300, bbox_inches='tight')
        plt.show()


```
