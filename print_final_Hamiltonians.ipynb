```python
import os
import sys
import sciexplorer
import json
from quantum_gs_final_runs.physics_run_quantumspins import models
# Set path to examples folder
EXAMPLES_PATH="/Users/mnaegel/Documents/sciexplorer_framework/sciexplorer/examples"
sys.path.append(EXAMPLES_PATH)
from simulators.quantum_spins import QuantumSpinSystem
import jax.numpy as jnp
```

```python
# get list of a json files in subdirectorires
exp_files = []
for root, dirs, files in os.walk('./quantum_gs_final_runs/'):
    for file in files:
        p = os.path.join(root, file)
        if '/run_' in p:
            exp_files.append(p)
for root, dirs, files in os.walk('./quantum_dynamics_final_runs/'):
    for file in files:
        p = os.path.join(root, file)
        if '/run_' in p:
            exp_files.append(p)
```

```python
time_list = []
for i, f in enumerate(exp_files):
    arguments = None
    with open(f, 'r') as file:
        exp_data = json.load(file)
    for msg in exp_data['messages']:
        tool_calls = msg.get('tool_calls', [])
        for call in tool_calls:
            function_name = call.get('function', {}).get('name', '')
            if function_name == 'announce_Hamiltonian':
                arguments = json.loads(call.get('function', {}).get('arguments', {}))
    exp_type = f.split('/')[-2]
    #print(f"File {i}/{len(exp_files)}: {f}, Hamiltonian: {arguments.get('Hamiltonian', '')}")
    if 'old' in exp_type or 'new' in exp_type:
        continue
    H_true = models[exp_type]['hamiltonian_code']
    H_pred = arguments.get('Hamiltonian', '')
    print(f"\n{exp_type}------------------------------------------------")
    print(f"error: {exp_data['result'].get('error', 'no error reported')}")
    print(f"log_fidelity/N {exp_data['result']['fidelity']}, fidelity^(1/N) {jnp.exp(exp_data['result']['fidelity'])}, file: {f}")
    print(f"H_true--------------------")
    print(f"{H_true}")
    print(f"H_pred--------------------")
    print(f"{H_pred}")
```
